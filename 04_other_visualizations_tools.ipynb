# Introduction to Seaborn

What is Seaborn?
Seaborn is a powerful Python data visualization library built on top of Matplotlib. It provides a high-level interface for creating attractive and informative statistical graphics.

Seaborn is often chosen over Matplotlib because it simplifies complex visualizations and provides better aesthetics by default. Here are the key reasons why Seaborn is preferred in many cases:

1. Simplicity & Less Code
2. Better Default Aesthetics
3. Built-in Statistical Features
4. Handles Categorical Data Easily
5. Easy Multi-Plot Grids

When to Use Matplotlib Instead of Seaborn?

## 0. Loading Library and Data

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Set the style
sns.set_style("whitegrid")
```

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("sakshigoyal7/credit-card-customers")

print("Path to dataset files:", path)

# Load the dataset into a pandas DataFrame
df = pd.read_csv(path + "/BankChurners.csv")  

# Display the first few rows of the dataset
print(df.head(3))
```

## 1. Basic Plots: Scatter Plot, Box Plot, Histogram, Line Plot

### A. Scatter Plot

```python
# Create a scatter plot
sns.scatterplot(x="Credit_Limit", y="Total_Revolving_Bal", data=df)
plt.title("Credit Limit vs Revolving Balance")
plt.show()
```

### B. Box Plot

```python
sns.boxplot(x="Credit_Limit", y="Income_Category", data=df)
plt.title("Distribution of Credit Limit by Income_Category")
plt.show()
```

### C. Histogram

```python
sns.histplot(x="Credit_Limit",data=df, bins=10, color='skyblue', edgecolor='black')
plt.title("Distribution of Credit Limit", fontsize=14, fontweight='bold')
```

### D. Line Plot

```python
sns.lineplot(x="Customer_Age",y="Credit_Limit", data=df)
plt.title("Distribution of Credit Limit by Customer Age", fontsize=14, fontweight='bold')
```

### E. Bar Plots

```python
sns.barplot(x="Income_Category", y="Credit_Limit", data=df, palette="rocket")
plt.title(" Credit Limit by Income Category")
plt.xlabel("Income Category")
plt.ylabel(" Credit Limit")
plt.xticks(rotation=45)
plt.show()
```

#### E. 1 Bar Graph with Grouping

- In the given grouped bar plot, the data is being grouped in two ways:

    - Primary grouping: By "Income_Category" (x-axis)

    - Secondary grouping: By "Gender" (hue parameter)

The bars are grouped first by Income_Category, and within each category, they are further divided by Gende

```python
sns.barplot(x="Income_Category", y="Credit_Limit", hue="Gender", data=df, palette="viridis")
plt.title("Credit Limit by Income Category and Gender")
plt.xlabel("Income Category")
plt.ylabel("Average Credit Limit")
plt.xticks(rotation=45)
plt.legend(title="Gender")
plt.show()
```
