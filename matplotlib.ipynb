```python
import pandas as pd
```

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("sakshigoyal7/credit-card-customers")

print("Path to dataset files:", path)
```

```python
# Load the dataset into a pandas DataFrame
df = pd.read_csv(path + "/BankChurners.csv")  

# Display the first few rows of the dataset
print(df.head())
```

### Lesson 1: Introduction to Matplotlib and Basic Line Plot

Objective: By the end of this lesson, you'll be able to:
- Import Matplotlib
- Create a simple line plot
- Add labels and a title to your plot

```python
# Step 0: Import Matplotlib
import matplotlib.pyplot as plt

# Step 1: Prepare the data
ages = df['Customer_Age']
credit_limits = df['Credit_Limit']

# Create a line plot
plt.figure(figsize=(10, 6))
plt.scatter(ages, credit_limits)

# Add labels and title
plt.xlabel('Customer Age')
plt.ylabel('Credit Limit')
plt.title('Credit Limit vs. Customer Age')

# Display the plot
plt.show()
```

```python
# Repeate#1 
import matplotlib.pyplot as plt

# prepare data
education_lvl_unique = df["Education_Level"].unique()
education_lvl_unique.sort()

credit_limits = df["Credit_Limit"]
avg_credit_limits = df.groupby("Education_Level")["Credit_Limit"].mean() #calculate the average credit limit for each education level


# plot data
plt.figure(figsize=(10,6))
plt.plot(education_lvl_unique, avg_credit_limits)

#lable plots
plt.xlabel("Education Level")
plt.ylabel("Credit Limit")
plt.title("Education vs Credit Limit")

#Display
plt.show()
```

### Lesson 2: Customizing Plots
Objective: By the end of this lesson, you'll be able to:
- Change the colors of your plots13.
- Adjust the line style (e.g., solid, dashed).
- Add markers to your line plots or scatter plots

```python
import matplotlib.pyplot as plt

# Prepare Data
ages = df['Customer_Age'].unique()
ages.sort()
avg_credit_limits = df.groupby('Customer_Age')['Credit_Limit'].mean()

# Plot 
plt.figure(figsize=(10,6))
plt.plot(ages, avg_credit_limits,
         color='green',           # Change line color
         linestyle='--',           # Change to dashed line
         marker='o',               # Add circular markers
         markersize=8,             # Increase marker size
         markerfacecolor='yellow', # Change marker fill color
         markeredgecolor='red'    # Change marker edge color
         )
# Lable
plt.xlabel("Age", fontsize=12)
plt.ylabel("Avg Credit Limit", fontsize=12)
plt.title("Average Credit Limit by Education Level", fontsize=14, fontweight='bold')

#Show 
plt.tight_layout()  # Adjust layout to prevent cutting off labels
plt.show()
```

```python
import matplotlib.pyplot as plt

# Prepare data
education_lvl_unique = df["Education_Level"].unique()
education_lvl_unique.sort()
avg_credit_limits = df.groupby("Education_Level")["Credit_Limit"].mean()

# Customized plot with color, line style, and markers
plt.figure(figsize=(12,6))  # Increased figure size for better visibility
plt.plot(education_lvl_unique, avg_credit_limits, 
         color='purple',           # Change line color
         linestyle='--',           # Change to dashed line
         marker='o',               # Add circular markers
         markersize=8,             # Increase marker size
         markerfacecolor='yellow', # Change marker fill color
         markeredgecolor='red')    # Change marker edge color

# Customize labels and title
plt.xlabel("Education Level", fontsize=12, fontweight='bold')
plt.ylabel("Average Credit Limit", fontsize=12, fontweight='bold')
plt.title("Average Credit Limit by Education Level", fontsize=14, fontweight='bold')

# Customize axis
plt.ylim(0, max(avg_credit_limits) * 1.1)  # Set y-axis to start at 0 and end a bit above the max value
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability

# Add grid for easier reading
plt.grid(True, linestyle=':', alpha=0.7)

# Display the plot
plt.tight_layout()  # Adjust layout to prevent cutting off labels
plt.show()
```
