# 2. Plotting Numerical Values

### Overview
In this section, we'll explore how to visualize numerical data using Matplotlib. We'll cover:

- **2.a Individual Plots**: Creating a single, standalone plot.
- **2.b Comparing Two Plots**: Displaying two related plots side-by-side for comparison.
- **2.c Multiple Plots**: Plotting one plot a time is not practical when working with large data sets. Plotting with loops

### Plotting Pattern: A Step-by-Step Guide

1.  **Prepare Data**: Clean, organize, and prepare your numerical data for plotting. This might involve loading data, filtering, or transforming it.

2.  **Plot Data**: Use Matplotlib functions (e.g., `plt.plot()`, `plt.hist()`, `plt.scatter()`) to create the desired plot type from your data.

3.  **Label Data**: Enhance clarity and understanding by adding titles, axis labels (x and y), legends, and annotations to your plot.

4.  **Display/Save Data**: Render the plot to the screen using `plt.show()` or save it to a file using `plt.savefig()`.

###  **2.a Individual Plots**: Creating a single, standalone plot.


```python
# Prepare Data 
credits_limit = df["Credit_Limit"]

# Plot Data
plt.hist(credits_limit,bins=10)

# Lable Plot
plt.title("Distribution of Credit Limit")
plt.xlabel("Credit Limit")
plt.ylabel("Frequency")

# Display data
plt.tight_layout()
plt.show()
```

```python
# Prepare Data 
Avg_Utilization_Ratio = df["Avg_Utilization_Ratio"]

# Plot Data
plt.boxplot(Avg_Utilization_Ratio)

# Lable Plot
plt.title("Box Plot of Avg Utilization Ratio")
plt.xlabel("Avg Utilization Ratio")

# Display data
plt.tight_layout()
plt.show()
```

### **2.b Comparing Two Plots**: Displaying two related plots side-by-side for comparison.


```python
# Prepare Data 
credits_limit = df["Credit_Limit"]
revolving_balance = df["Total_Revolving_Bal"]


# Plot Data
plt.scatter(credits_limit,revolving_balance, alpha=0.7)

# Lable Plot
plt.title('Scatter Plot: Relationship Between Two Variables')
plt.xlabel("Credit Limit")
plt.ylabel("Revolving Balance")

# Display data
plt.tight_layout()
plt.show()
```

### **2.c Multiple Plots**: Plotting one plot a time is not practical when working with large data sets. Plotting with loops

```python
# Preparing Data
numeric_columns = df.select_dtypes(include=["number"]).columns

# Ploting Data
for columns in numeric_columns:
    plt.figure()
    plt.hist(df[columns],bins=10)
    # Labeling Plots
    plt.title(f'Histogram of {columns}')
    plt.xlabel(columns)
    plt.ylabel('Frequency')

# Display Plots
plt.tight_layout()
plt.show

```

# 2. Plotting Categorical Variables

## Overview
In this section, we’ll explore how to visualize categorical data using Matplotlib. We’ll cover:

- **2.a Individual Plots**: Creating a single, standalone plot for categorical data.
- **2.b Comparing Two Plots**: Displaying two related plots side-by-side for comparison.
- **2.c Multiple Plots**: Automating the creation of multiple categorical plots using loops.

## Plotting Pattern: A Step-by-Step Guide

1. **Prepare Data**: Clean, organize, and prepare your categorical data for plotting. This might involve grouping, summarizing, or transforming the data into a suitable format (e.g., dictionaries or DataFrames).

2. **Plot Data**: Use Matplotlib functions (e.g., `plt.bar()`, `plt.scatter()`, `plt.plot()`) to create visualizations from your categorical data. For example:



```python
# Identifying categorical variables 
df.select_dtypes(include=["object"]).columns
```

###  **2.a Individual Plots**: Creating a single, standalone plot.


```python
# Does the business have more male customers? 

# Prepare Data
gender_counts = df['Gender'].value_counts()

# Plot Data
plt.bar(gender_counts.index, gender_counts.values)

# Lable Plot
plt.title("Count of male and female customers")
plt.xlabel("Male vs. Female")
plt.ylabel("Count")

# Display Plot
plt.tight_layout()
plt.show()
```

### **2.b Comparing Two Plots**: Displaying two related plots side-by-side for comparison.
- Rather than using matplotlib, we will use pandas package to keep bar graphs simpler

```python
# Preparing data

#analyze the relationship between two or more categorical variables by creating a table that shows how the variables intersect
gender_marital = pd.crosstab(df['Gender'], df['Marital_Status'])

# Ploting Data
gender_marital.plot(kind='bar', figsize=(8, 6))

# Labeling Data 
plt.title('Gender vs. Marital Status')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.xticks(rotation=0)  # Keep x-axis labels horizontal

# Displaying Data
plt.tight_layout()
plt.show()
```

```python
#analyze the relationship between two or more categorical variables by creating a table that shows how the variables intersect
gender_marital_education = pd.crosstab(df['Gender'], [df['Marital_Status'], df['Education_Level']])

# Ploting Data
gender_marital_education.plot(kind='bar', figsize=(12, 6))

plt.legend(title='Marital Status, Education Level', bbox_to_anchor=(1.05, 1), loc='upper left')
```

### **2.c Multiple Plots**: Plotting one plot a time is not practical when working with large data sets. Plotting with loops

```python
categories = ['Gender', 'Marital_Status']
for category in categories:
    plt.figure(figsize=(8, 6))
    counts = df[category].value_counts()
    plt.bar(counts.index, counts.values)

    plt.tight_layout()
    plt.show()
```
